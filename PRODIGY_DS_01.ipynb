IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

CREATING SAMPLE DATASET

```python
data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Male", "Female", "Female", "Male", "Female"],
    "Age": [25, 31, 42, 35, 28, 40, 29, 32, 45, 38]
}

df = pd.DataFrame(data)
```

BAR CHART FOR GENDER (Catogorical variable)

```python
# Bar chart for Gender
df["Gender"].value_counts().plot(kind="bar", figsize=(8, 6))
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

HISTOGRAM FOR AGE (continuous variable) 

```python
# Histogram for Age
df["Age"].plot(kind="hist", bins=10, edgecolor="black", figsize=(8, 6))
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
